```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

FS = 15 #Fontsize of caption
```

## 3.1 Adding 2 images

```python
#Read the image test
img_a = cv2.imread('prarie.jpg')
img_b = cv2.imread('giraffe.jpg')
img_b_resize = cv2.resize(img_b, (img_a.shape[1], img_a.shape[0]))
img_c = cv2.add(img_a, img_b_resize)

plt.figure(figsize = (15, 5))

#Plot images
plt.subplot(1, 3, 1)
plt.imshow(img_a)
plt.title('Image 1', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_b)
plt.title('Image 2', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_c)
plt.title('Combination image', fontsize = FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```

## 3.2 Multiplying 2 or more images

```python
#Read the image test with range [0, 1]
img_a = cv2.imread('earth1.jpg') / 255.0
img_b = cv2.imread('earth2.jpg') / 255.0
img = np.multiply(np.multiply(img_a, img_b), np.multiply(img_a, img_b))

plt.figure(figsize = (10, 5))

#Plot images
plt.subplot(1, 2, 1)
plt.imshow(img_a)
plt.title('2D Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img)
plt.title('3D Image', fontsize = FS)
plt.axis('off')
```

## 3.3 Dividing 2 images

```python
#Read the image test with range [0, 1]
img_noText = cv2.imread('gradient.jpg', cv2.IMREAD_GRAYSCALE) / 255.0
img_haveText = cv2.imread('gradient_with_text.jpg', cv2.IMREAD_GRAYSCALE) / 255.0
img_divide = cv2.divide(img_haveText, img_noText + 1e-8)
text = img_divide > 1

plt.figure(figsize = (15, 5))

#Plot images
plt.subplot(1, 3, 1)
plt.imshow(img_noText, cmap = 'gray', vmin = 0, vmax = 1)
plt.title('No text Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_haveText, cmap = 'gray', vmin = 0, vmax = 1)
plt.title('Have text Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(text, cmap = 'gray', vmin = 0, vmax = 1)
plt.title('Text Image', fontsize = FS)
plt.axis('off')
```

## 3.4 Geometric operations

```python
#Read the image test
image_1 = cv2.imread('bird.jpg')  #Original image

#1. Flipping an image up-down and left-right
flip_ud = cv2.flip(image_1, 0)  #Flip up-down
flip_lr = cv2.flip(image_1, 1)  #Flip left-right

#2. Rotating an image with angle theta in degree
image_2 = cv2.imread('eight.png')
theta = 90  #Angle in degrees
(h, w) = image_2.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, theta, 1.0)
rotated_image = cv2.warpAffine(image_2, rotation_matrix, (w, h))

#3. Cropping an image
image_3 = cv2.imread('mandrill.tif')
x1, x2, y1, y2 = 150, 450, 100, 500  #Crop coordinates
cropped_image = image_3[y1:y2, x1:x2]

#4. Zooming an image with different interpolation methods
zoom_bicubic = cv2.resize(image_1, None, fx = 2, fy = 2, interpolation = cv2.INTER_CUBIC)
zoom_nearest = cv2.resize(image_1, None, fx = 2, fy = 2, interpolation = cv2.INTER_NEAREST)
zoom_bilinear = cv2.resize(image_1, None, fx = 2, fy = 2, interpolation = cv2.INTER_LINEAR)

#Plot results
#1. Flipping results
plt.figure(figsize = (10, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(flip_ud, cv2.COLOR_BGR2RGB))
plt.title('Flip Up-Down', fontsize = FS)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(flip_lr, cv2.COLOR_BGR2RGB))
plt.title('Flip Left-Right', fontsize = FS)
plt.axis('off')

plt.tight_layout()
plt.show()

#2. Rotated image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image', fontsize = FS)
plt.axis('off')

plt.tight_layout()
plt.show()

#3. Cropped image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_3, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image', fontsize = FS)
plt.axis('off')

plt.tight_layout()
plt.show()

# Zooming results
plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize = FS)
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(zoom_bicubic, cv2.COLOR_BGR2RGB))
plt.title('Bicubic Zoom', fontsize = FS)
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(zoom_nearest, cv2.COLOR_BGR2RGB))
plt.title('Nearest Zoom', fontsize = FS)
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(zoom_bilinear, cv2.COLOR_BGR2RGB))
plt.title('Bilinear Zoom', fontsize = FS)
plt.axis('off')

plt.tight_layout()
plt.show()
```
